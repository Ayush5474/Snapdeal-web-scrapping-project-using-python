```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://www.snapdeal.com/products/mens-footwear?sort=plrty")
page
```

```python
soup = BeautifulSoup(page.content,"html.parser")
soup
```

```python
period = soup.find(class_="product-title")
```

```python
period
```

```python
period_tags = soup.select(".product-title")
period_tags
```

```python
period_tags = soup.select(".product-title")
period_tags
import re
name_list=[]
for i in range(0,len(period_tags)):
    text = period_tags[i].get_text()
    text_1=re.split("\n",text)
    a=text_1[0]
    name_list.append(a)
name_list
```

```python
period_tags = soup.select(".lfloat.product-price")
period_tags

import re
price_list=[]
for i in range(0,len(period_tags)):
    text = period_tags[i].get_text()
    text_1=re.findall("[0-9]+",text)
    a=text_1[0]
    price_list.append(a)

price_list

for p in price_list:
    print("Rs.",p)
```

```python
period_tags = soup.select(".product-discount")
period_tags

import re
discount_list=[]
for i in range(-1,len(period_tags)):
    text = period_tags[i].get_text()
    text_1=re.findall("[0-9]+",text)
    a=text_1[0]
    discount_list.append(a)

discount_list

for D in discount_list:
    print(D,"% Off")
```

```python
len(name_list)
```

```python
period_tags = soup.select(".lfloat.product-desc-price.strike ")
period_tags

import re
Off_price_list=[]
for i in range(0,len(period_tags)):
    text = period_tags[i].get_text()
    text_1=re.findall("[A-Z a-z]+[0-9]+",text)
    a=text_1[0]
    Off_price_list.append(a)

Off_price_list

for O in Off_price_list:
    print(O)
```

```python
final_one = {"product name":name_list,"discount percentage":discount_list, "price":price_list,"Off price":Off_price_list}
final_one
```

```python
dict = pd.DataFrame(final_one)
dict
```
